```python
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
df
```

```python
inputs = df.drop('salary_more_then_100k', axis = 'columns')
inputs.head()
```

```python
target = df['salary_more_then_100k']
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
company_ = le_company.fit_transform(inputs['company'])
inputs = inputs.assign(company_n = company_)

job_ = le_job.fit_transform(inputs['company'])
inputs = inputs.assign(job_n = job_)

degree_ = le_degree.fit_transform(inputs['company'])
inputs = inputs.assign(degree_n = degree_)
```

```python
inputs
```

```python
inputs_n = inputs.drop(['company', 'job', 'degree'], axis = 'columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```

```python
model.predict([[2,2,1]])
```
